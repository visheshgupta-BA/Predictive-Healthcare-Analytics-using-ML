## Model Building and Model Selection

```python
df
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:,:-1]
y = df['Dataset']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=123)
```

```python
## Performing scaling to get the distribution of data in a relevant shape
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#!pip install imblearn
```

```python
## performing sampling techniques to make the data balanced
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = sm.fit_resample(X_train_scaled, y_train)
```

## LR (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix

logreg = LogisticRegression()

logreg.fit(X_train_scaled, y_train)

predictions_train = logreg.predict(X_train_scaled)
predictions_test = logreg.predict(X_test_scaled)

logreg_acc_train = accuracy_score(y_train, predictions_train)
logreg_acc_test = accuracy_score(y_test, predictions_test)

print("Training Accuracy of the Logistic Regression Model is: ", logreg_acc_train)
print("Testing Accuracy of the Logistic Regression Model is: ", logreg_acc_test)
print('\n')

print(classification_report(y_test, predictions_test))

print('Confusion matrix')
print('-------------------------------------------------------')
print(confusion_matrix(y_test, predictions_test))
```

## KNN

```python
# Building a model for K from 1-10

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

accuracy_val_k = []

k_val = list(range(1,11))

for k in k_val:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train_scaled, y_train)
  y_pred_val = knn.predict(X_test_scaled)
  accuracy = accuracy_score(y_test, y_pred_val)
  accuracy_val_k.append(accuracy)


accuracy_val_k
```

```python
from sklearn.neighbors import KNeighborsClassifier

# K-Neighbors with best K=1
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train_scaled, y_train)

predictions_train = knn.predict(X_train_scaled)
predictions_test = knn.predict(X_test_scaled)

knn_acc_train = accuracy_score(y_train, predictions_train)
knn_acc_test = accuracy_score(y_test, predictions_test)


print("Training Accuracy of the KNN Model is: ", knn_acc_train)
print("Testing Accuracy of the KNN Model is: ", knn_acc_test)
print('\n')

print(classification_report(y_test, predictions_test))

print('Confusion matrix')
print('-------------------------------------------------------')
print(confusion_matrix(y_test, predictions_test))
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(max_depth=4, random_state=42)

# Train the Decision Tree Classifier on the training data
dt_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test data
y_train_hat = dt_classifier.predict(X_train_scaled)
y_test_hat = dt_classifier.predict(X_test_scaled)


dt_acc_train = accuracy_score(y_train, y_train_hat)
dt_acc_test = accuracy_score(y_test, y_test_hat)


print("Training Accuracy of the Decision Tree Model is: ", dt_acc_train)
print("Testing Accuracy of the Decision Tree Model is: ", dt_acc_test)
print('\n')

print(classification_report(y_test, y_test_hat))

print('Confusion matrix')
print('-------------------------------------------------------')
print(confusion_matrix(y_test, y_test_hat))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc1 = RandomForestClassifier(n_jobs=-1,random_state=123)
rfc1.fit(X_train_scaled, y_train)

y_train_hat = rfc1.predict(X_train_scaled)
y_test_hat = rfc1.predict(X_test_scaled)

rf_acc_train = accuracy_score(y_train, y_train_hat)
rf_acc_test = accuracy_score(y_test, y_test_hat)

print("Training Accuracy of the Random-Forest Model is: ", rf_acc_train)
print("Testing Accuracy of the Random-Forest Model is: ", rf_acc_test)
print('\n')

print(classification_report(y_test, y_test_hat))

print('Confusion matrix')
print('-------------------------------------------------------')
print(confusion_matrix(y_test, y_test_hat))
```

## Tuning the Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Instantiate the Random Forest Classifier
rf_classifier = RandomForestClassifier()

# Define the hyperparameter grid for tuning
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of trees in the forest
    'max_depth': [None, 10, 20],  # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4]  # Minimum number of samples required to be at a leaf node
}

# Perform grid search using 5-fold cross-validation
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train)  # Replace X_train and y_train with your training data

# Print the best hyperparameters and the corresponding accuracy
print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Accuracy: ", grid_search.best_score_)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate the Random Forest Classifier with the best hyperparameters
rfc = RandomForestClassifier(max_depth=None, min_samples_leaf=1, min_samples_split=10, n_estimators=50)

rfc.fit(X_train_scaled, y_train)

y_train_hat = rfc.predict(X_train_scaled)
y_test_hat = rfc.predict(X_test_scaled)

rf_acc_train = accuracy_score(y_train, y_train_hat)
rf_acc_test = accuracy_score(y_test, y_test_hat)

print("Training Accuracy of the Random-Forest Model is: ", rf_acc_train)
print("Testing Accuracy of the Random-Forest Model is: ", rf_acc_test)
print('\n')

print(classification_report(y_test, y_test_hat))

print('Confusion matrix')
print('-------------------------------------------------------')
```

## Tuning the SVM

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define the hyperparameter grid to search
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto'] + [0.1, 1, 10]
}

# Instantiate the SVM classifier
svm = SVC()

# Create GridSearchCV object with cross-validation
grid_search = GridSearchCV(svm, param_grid, cv=5)

# Fit the GridSearchCV object to the training data
grid_search.fit(X_train_scaled, y_train)  # Replace X_train_scaled and y_train with your training data

# Get the best hyperparameters and model
best_params = grid_search.best_params_
best_svm = grid_search.best_estimator_

# Fit the best model to the training data
best_svm.fit(X_train_scaled, y_train)

# Make predictions on the training and test data
predictions_train = best_svm.predict(X_train_scaled)
predictions_test = best_svm.predict(X_test_scaled)

# Calculate accuracy
svm_acc_train = accuracy_score(y_train, predictions_train)
svm_acc_test = accuracy_score(y_test, predictions_test)

# Print results
print("Best Hyperparameters: ", best_params)
print("Training Accuracy of the SVM Model: ", svm_acc_train)
print("Testing Accuracy of the SVM Model: ", svm_acc_test)
```
